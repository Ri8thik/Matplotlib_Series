```python
import matplotlib.pyplot as plt
```

```python
classes=["Python","R","AI","ML","DS"]
class1_students=[45,15,10,15,15]
```

```python
explode=[0,0.1,0,0.1,0]
textprops={"fontsize":15}
plt.pie(class1_students,labels=classes,explode=explode,autopct="%0.1f%%",shadow=True,radius=1.9,textprops=textprops)
plt.show()
```

```python
explode=[0,0.1,0,0.1,0]
textprops={"fontsize":15}
wedgeprops={"linewidth":4,"width":1,"edgecolor":"k"}
plt.pie(class1_students,
        labels=classes,
        explode=explode,
        autopct="%0.1f%%",
        shadow=True,
        radius=1,
        textprops=textprops,
        pctdistance=0.6,
        labeldistance=1.1,
        counterclock=True,
        wedgeprops=wedgeprops,
        center=(2,3),
        frame=True,
        rotatelabels=False)
plt.show()
```
