```python
import matplotlib.pyplot as plt 
import numpy as np 
from matplotlib import style 
```

```python
classes =["Python","R","AI","ML","DS"]
class1_student =[30,10,20,30,10]
class2_student =[40,10,20,20,10]
class3_student =[35,5,30,15,15]
```

```python
width=0.2
plt.bar(classes,class1_student,width,color="g",label="Class1_students")
plt.title("All student bar chat ",fontsize=30)
plt.xlabel("Subjects",fontsize=20)
plt.ylabel("No of students",fontsize=20)
plt.legend()
plt.show()
```

```python
width=0.2
plt.barh(classes,class1_student,width,color="g",label="Class1_students")
plt.title("All student bar chat ",fontsize=30)
plt.xlabel("Subjects",fontsize=20)
plt.ylabel("No of students",fontsize=20)
plt.legend()
plt.show()
```

```python
width=0.2
plt.figure(figsize=(16,9))
plt.bar(classes,class1_student,width,color="g",label="Class1_students")
plt.title("All student bar chat ",fontsize=30)
plt.xlabel("Subjects",fontsize=20)
plt.ylabel("No of students",fontsize=20)
plt.legend()
plt.show()
```

```python
width=0.2
plt.figure(figsize=(16,9))
style.use("ggplot")
class_index=np.arange(len(classes))

plt.bar(class_index,class1_student,width,color="g",label="Class1_students")
plt.bar(class_index+width,class2_student,width,color="b",label="Class2_students")
plt.bar(class_index+width+width,class3_student,width,color="y",label="Class3_students")
plt.xticks(class_index+width,classes)
plt.title("All student bar chat ",fontsize=30)
plt.xlabel("Subjects",fontsize=20)
plt.ylabel("No of students",fontsize=20)
plt.grid(color="k",linestyle="--",linewidth=1)
plt.legend()
plt.show()
```

```python
width=0.2
plt.figure(figsize=(16,9))
style.use("ggplot")
class_index=np.arange(len(classes))

plt.barh(class_index,class1_student,width,color="g",label="Class1_students")
plt.barh(class_index+width,class2_student,width,color="b",label="Class2_students")
plt.barh(class_index+width+width,class3_student,width,color="y",label="Class3_students")
plt.yticks(class_index+width,classes)
plt.title("All student bar chat ",fontsize=30)
plt.ylabel("Subjects",fontsize=20)
plt.xlabel("No of students",fontsize=20)
plt.grid(color="k",linestyle="--",linewidth=1)
plt.legend()
plt.show()
```
