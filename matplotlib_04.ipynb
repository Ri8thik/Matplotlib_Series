```python
from matplotlib import pyplot as plt
import numpy as np 
from matplotlib import style 
```

```python
classes =["Python","R","AI","ML","DS"]
class1_student =[30,10,20,30,10]
class2_student =[40,10,20,20,10]
class3_student =[35,5,30,15,15]
```

```python
plt.bar(classes,class1_student)
plt.show()
```

```python
plt.barh(classes,class1_student)
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.2,align="edge",color="g",edgecolor="k",linewidth=3)
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.2,align="edge",color="g",edgecolor="k",linewidth=3,alpha=0.6)
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.2,align="edge",color="g",edgecolor="k",linewidth=2,label="Class_1 student ",
        linestyle="--")
plt.legend()
plt.show()
```

```python
plt.bar(classes,class1_student,width=0.2,align="edge",color="g",edgecolor="k",linewidth=2,label="Class_1 student ",
        linestyle="--",visible=False) # visible will clear the bar if it will take the False value and bydefalt it will take the True value 
plt.legend()
plt.show()
```
