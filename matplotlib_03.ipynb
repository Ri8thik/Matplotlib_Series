```python
from matplotlib import pyplot as plt 
import numpy as np 
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML student age histogram")
plt.xlabel("ml student age")
plt.ylabel("NO of student ")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",orientation="horizontal",color="g",label="ML student")
plt.title("ML student age histogram")
plt.xlabel("ml student age")
plt.ylabel("NO of student ")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",color="g",label="ML student")
plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",color="b",label="py student")
plt.title("ML student age histogram")
plt.xlabel("ml student age")
plt.ylabel("NO of student ")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",color=["g","b"],
         label=["ML student","PY student"])
# plt.hist(py_student_age,bins,rwidth=0.8,histtype="bar",orientation="vertical",color="b",label="py student")
plt.title("ML student age histogram")
plt.xlabel("ml student age")
plt.ylabel("NO of student ")
plt.legend()
plt.show()
```

```python
from matplotlib import style 
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",color=["g","b"],
         label=["ML student","PY student"])

plt.title("ML student age histogram")
plt.xlabel("ml student age")
plt.ylabel("NO of student ")
plt.legend()
plt.show()
```

```python
from matplotlib import style 
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="vertical",color=["g","b"],
         label=["ML student","PY student"])

plt.title("ML student age histogram")
plt.xlabel("ml student age")
plt.ylabel("NO of student ")
plt.grid(color="k")
plt.legend()
plt.show()
```
