```python
import matplotlib.pyplot as plt
```

```python
classes=["Python","R","AI","ML","DS"]
class1_students=[45,15,10,15,15]
```

```python
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
explode=[0,0,0.1,0,0.1]
plt.pie(class1_students,labels=classes,explode=explode)
plt.show()
```

```python
explode=[0,0.1,0.1,0,0.1]
colors=["r","g","b","y","k"]
plt.pie(class1_students,labels=classes,explode=explode,colors=colors)
plt.show()
```

```python
explode=[0,0,0.1,0,0.1]
colors=["r","g","b","y","orange"]
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%")
plt.show()
```

```python
explode=[0,0.1,0.1,0,0.1]
colors=["r","g","b","y","orange"]
textprops={"fontsize":15}
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.1f%%",shadow=True,radius=1.8,startangle=90,
       textprops=textprops)
plt.show()
```
