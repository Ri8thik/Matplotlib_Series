```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Dataset.csv")
data.shape
```

```python
data
```

```python
plt.figure(figsize=(16,9))
plt.scatter(data["Class"],data["Study Hrs"])
plt.title("Student data set")
plt.xlabel("class")
plt.ylabel("Study Hrs")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(data["Class"],data["Study Hrs"],c="g",marker="*",s=1000)
plt.title("Student data set")
plt.xlabel("class")
plt.ylabel("Study Hrs")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(data["Class"],data["Study Hrs"],c="g",marker="*",s=1000,alpha=0.6,linewidth=10,edgecolor="b")
plt.title("Student data set")
plt.xlabel("class")
plt.ylabel("Study Hrs")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.scatter(data["Class"],data["Study Hrs"],c="g",marker="*",s=100,alpha=0.6,linewidth=10,edgecolor="b",label="Study Hrs")
plt.scatter(data["Class"],data["Percentege"],c="k",marker="o",s=100,alpha=0.6,linewidth=10,edgecolor="k",label="Percentege")
plt.title("Student data set")
plt.xlabel("class")
plt.ylabel("Study Hrs and Percentege")
plt.legend()
plt.show()
```
